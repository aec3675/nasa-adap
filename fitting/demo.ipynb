````python
# 1-D GP fit takes kernel and SNCollection

# DO LENGTH SCALE / BOUNDS NEED TO BE DEFINED IN LOG IF UUSING LOG-SPACING?
kernel = MaternKernel(length_scale=10, length_scale_bounds=(1e-10, 10.), nu=1.5).kernel + WhiteNoiseKernel(1., (1e-10, 10.)).kernel

gp = GP(sne_collection=IIb, kernel=kernel)
````

````python
# perform fit for one filter
gp.predict_gp(filt='g', phasemin=-20, phasemax=50, test_size=0.9, plot=True)
````
